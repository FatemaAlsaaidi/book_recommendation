##  To Apply clustering algorithm to be able to classification the user interesting of books we need to convert final_rating to pivot table

```python
book_pivot = final_rating.pivot_table(columns ='user_id', index = 'title', values ='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
#handle with NaN values
book_pivot.fillna(0, inplace = True)
```

```python
book_pivot
```

```python
# to enter book_pivot to the model we need to apply some techniqes to ignore th 0 value to ignor to claculate to not take more time the clustering algorithm becuse 0 value is already nothing 
from scipy.sparse import csr_matrix
```

```python
book_sparse =csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
#print 6 book id whose are similarity
distance
```

```python
#print 6 book index whose are similarity
suggestion
```

```python
# print those name of 6 books whose are similarity
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

### Now we see how our model recommending 

```python
#np.where(book_pivot.index=="2nd Chance")[0][0]
```

```python
def recommend_book(book_name):
    #get the index
    book_id  = np.where(book_pivot.index==book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = 'A Bend in the Road'
```

```python
recommend_book(book_name)
```
